**LINEAR REGRESSION**

```python
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.model_selection import train_test_split # For splitting dataset
from sklearn.metrics import mean_absolute_error, accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix, multilabel_confusion_matrix # Evaluation metrics
from sklearn.pipeline import make_pipeline # Creating a Pipeline though not required. Only for efficiency
from sklearn.preprocessing import StandardScaler # Scaling data
from sklearn.linear_model import LinearRegression # Linear Regression Model
from sklearn.metrics import mean_squared_error
```

```python
#correlation between the features and response
corr_mat = data.corr()
plt.subplots(figsize = (35,20))
sns.heatmap(corr_mat, annot = True, linewidths = 3, linecolor = 'black', cmap = 'Blues')
plt.title('Correlation Plot', size = 25)
plt.ylabel('Features', size = 20)
plt.xlabel('Features', size = 20)
plt.show()
```

```python
#splitting the feature and response variable
X = data.loc[:,['Year', 'Age', 'rotten_tomatoes', 'Netflix', 'Hulu',
       'Prime Video', 'Disney+', 'Multi-lingual', 'MultipleDirectors',
       'Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Family', 'Fantasy', 'FilmNoir', 'History',
       'Horror', 'Music', 'Musical', 'Mystery', 'News', 'RealityTV', 'Romance',
       'SciFi', 'Short', 'Sport', 'TalkShow', 'Thriller', 'War', 'Western', 'Runtime']].values
y = data.loc[:, ['IMDb']].values
```

```python
#converting dataframes to numpy arrays
X = np.array(X)
y = np.array(y)
```

```python
#standardizing and splitting into train and test sets
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = scaler.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
print("Shape of Feature Set:", X.shape)
print("Shape of Label:", y.shape)
```

```python
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
#making the regression model
model = make_pipeline(StandardScaler(), LinearRegression())
```

```python
model.fit(X_train, y_train) #fitting the model on the train set
```

```python
y_pred = model.predict(X_test) #testing the model on the test set
```

```python
# We can calculate the R^2 value
r_square = model.score(X_test, y_test)
print("R^2 =", r_square)
```

```python
# We can also calculate the mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error = ", mae)
```

```python
# We can also calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error = ", mse)
```

```python
residual = y_test - y_pred
```

```python
list_1 = []
for i in range(1, residual.shape[0]+1):
    list_1.append(i)

plt.figure(figsize = (9,6))
sns.distplot(residual, color ='blue')
plt.ylabel('Residual Density')
plt.xlabel('Relative Distance')
plt.title('Histogram of Test Residual Data')
plt.show()
```

```python
accuracy = model.score(X_test, y_test)
print("Accuracy of the linear regression model =", accuracy*100, "%")
```

**KNN ON NETFLIX**

```python
X1 = data.loc[:,['Year', 'Age', 'rotten_tomatoes', 'Hulu',
       'Prime Video', 'Disney+', 'Multi-lingual', 'MultipleDirectors',
       'Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Family', 'Fantasy', 'FilmNoir', 'History',
       'Horror', 'Music', 'Musical', 'Mystery', 'News', 'RealityTV', 'Romance',
       'SciFi', 'Short', 'Sport', 'TalkShow', 'Thriller', 'War', 'Western', 'Runtime', 'IMDb']].values
y1 = data.loc[:, ['Netflix']].values
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.20)
```

```python
scaler = StandardScaler()
scaler.fit(X1_train)
X1_train = scaler.transform(X1_train)
X1_test = scaler.transform(X1_test)
```

```python
print("Shape of Feature Set:", X1.shape)
print("Shape of Label:", y1.shape)
```

```python
print('Train set:', X1_train.shape,  y1_train.shape)
print('Test set:', X1_test.shape,  y1_test.shape)
```

```python
#making a knn classifier
classifier = KNeighborsClassifier(n_neighbors = 12)
classifier.fit(X1_train, y1_train)
```

```python
y1_pred = classifier.predict(X1_test)
```

```python
print(classification_report(y1_test, y1_pred))
```

```python
error = []
# Calculating error for K values between 1 and 40
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X1_train, y1_train)
    pred_i = knn.predict(X1_test)
    error.append(np.mean(pred_i != y1_test))
```

```python
plt.figure(figsize = (15, 6))
plt.plot(range(1, 40), error, color = 'red', linestyle = 'dashed', marker = 'o',
         markerfacecolor = 'blue', markersize = 10)
plt.title('Error Rate K Value for Netflix')
plt.xlabel('K Value')
plt.ylabel('Mean Error')

min_val = 1 + error.index(min(error))
print("Minimum error: ", np.around(min(error), decimals = 5),"at K =", min_val)
```

```python
accuracy = classifier.score(X1_test, y1_test)
print("Accuracy of the knn classifier on Netflix =", accuracy*100, "%")
```
